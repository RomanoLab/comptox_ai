```python
from comptox_ai.db.graph_db import GraphDB
import pandas as pd
from tqdm import tqdm_notebook
```

```python
db = GraphDB(hostname="neo4j.comptox.ai")
pf_db = pd.read_excel("/Users/jdr2160/Downloads/perfluorome database.xlsx")
```

```python
casrns = list(set([x.split(';')[0] for x in list(pf_db['CAS.RN'])]))
```

```python
len(casrns)
```

```python
count = 0
matches = []
unmatches = []

for crn in tqdm_notebook(casrns):
    res = db.run_cypher(f"MATCH (c:Chemical {{ xrefCasRN: \"{crn}\" }} ) RETURN c;")
    if len(res) > 0:
        count += 1
        matches.append(res[0]['c'])
    else:
        unmatches.append(crn)
        
print(count)
```
